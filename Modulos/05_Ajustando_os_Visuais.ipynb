```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tmdb = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/data-science-analise-exploratoria/main/Aula_0/tmdb_5000_movies.csv")
tmdb.head()
```

```python
total_de_outros_filmes_por_lingua = tmdb.query("original_language != 'en'")["original_language"].value_counts()
total_de_outros_filmes_por_lingua.head()
```

```python
sns.countplot(data=tmdb.query("original_language != 'en'"), x="original_language", order=total_de_outros_filmes_por_lingua.index, hue = "original_language")
```

```python
tmdb.query("original_language != 'en'")["original_language"].value_counts(normalize=True)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=tmdb.query("original_language != 'en'"), x="original_language", order=total_de_outros_filmes_por_lingua.index, hue = "original_language", palette="mako", hue_order=total_de_outros_filmes_por_lingua.index, stat = "percent")
plt.title("DistribuiÃ§Ã£o da lÃ­ngua original nos filmes exceto em inglÃªs")
plt.show()
```
